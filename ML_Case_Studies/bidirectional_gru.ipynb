```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from torch.nn.functional import softmax
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from tqdm.auto import tqdm
import os
import joblib

# Import EarlyStopping
from tensorflow.keras.callbacks import EarlyStopping

# --- Configuration and Setup ---
print("--- Checking for GPU setup ---")
if tf.config.list_physical_devices('GPU'):
    print("GPU found and configured for TensorFlow.")
    tf.config.experimental.set_memory_growth(tf.config.list_physical_devices('GPU')[0], True) # Dynamic memory allocation
else:
    print("No GPU found. Running on CPU. Consider installing TensorFlow with GPU support for faster training.")

# Define key parameters
SEQUENCE_LENGTH = 30
BATCH_SIZE = 32
MAX_EPOCHS = 200
EARLY_STOPPING_PATIENCE = 20
FINBERT_BATCH_SIZE = 64
MODEL_SAVE_DIR = 'stock_prediction_models_bidirectional_gru_enhanced_sentiment_reanalyzed' # New folder for reanalyzed sentiment
SCALER_SAVE_DIR = 'scalers_bidirectional_gru_enhanced_sentiment_reanalyzed' # New folder
MERGED_DATA_SAVE_FILE = 'merged_stock_sentiment_data_reanalyzed.csv' # New filename for merged data

# Create directories if they don't exist
os.makedirs(MODEL_SAVE_DIR, exist_ok=True)
os.makedirs(SCALER_SAVE_DIR, exist_ok=True)

# --- 1. Data Loading and Preprocessing ---
print("\n--- 1. Data Loading and Preprocessing ---")
try:
    news_df = pd.read_csv('aligned_news_data.csv')
    stock_df = pd.read_csv('aligned_stock_data.csv')
except FileNotFoundError:
    print("Ensure 'aligned_news_data.csv' and 'aligned_stock_data.csv' are in the same directory.")
    exit()

news_df['Date'] = pd.to_datetime(news_df['Date'], format='%d-%m-%Y')
stock_df['Date'] = pd.to_datetime(stock_df['Date'])

news_df = news_df.sort_values(by='Date').reset_index(drop=True)
stock_df = stock_df.sort_values(by='Date').reset_index(drop=True)

print("News Data Head:")
print(news_df.head())
print("\nStock Data Head:")
print(stock_df.head())

# --- 2. Optimized Sentiment Analysis with FinBERT (Re-analyze Headlines) ---
# This section explicitly re-runs the sentiment analysis on 'Headline'
print("\n--- 2. Optimized Sentiment Analysis with FinBERT (Re-analyzing Headlines) ---")

# Initialize FinBERT tokenizer and model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model_finbert = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert").to(device)
model_finbert.eval()

def get_finbert_sentiment_probabilities_batch(texts, batch_size=FINBERT_BATCH_SIZE):
    all_probabilities = []
    # Explicitly process the 'Headline' column data
    processed_texts = [
        str(t) if pd.notna(t) and t != '[Neutral] No news available for JPM' else ''
        for t in texts
    ]

    for i in tqdm(range(0, len(processed_texts), batch_size), desc="Processing News Sentiment"):
        batch_texts = processed_texts[i:i + batch_size]
        valid_texts_batch = [t for t in batch_texts if t]

        if not valid_texts_batch:
            all_probabilities.extend([[0.0, 0.0, 0.0]] * len(batch_texts))
            continue

        inputs = tokenizer(valid_texts_batch, return_tensors="pt", padding=True, truncation=True, max_length=512).to(device)
        with torch.no_grad():
            outputs = model_finbert(**inputs)
        probabilities = softmax(outputs.logits, dim=1).cpu().numpy()

        prob_idx = 0
        for text_original in batch_texts:
            if text_original == '':
                all_probabilities.append([0.0, 0.0, 0.0])
            else:
                all_probabilities.append([probabilities[prob_idx, 0], probabilities[prob_idx, 1], probabilities[prob_idx, 2]])
                prob_idx += 1
    return np.array(all_probabilities)

# Ensure to explicitly use news_df['Headline'].tolist()
finbert_probs = get_finbert_sentiment_probabilities_batch(news_df['Headline'].tolist(), batch_size=FINBERT_BATCH_SIZE)

# Assign new sentiment columns, effectively overwriting any previous ones if they existed
news_df['finbert_neutral_prob'] = finbert_probs[:, 0]
news_df['finbert_positive_prob'] = finbert_probs[:, 1]
news_df['finbert_negative_prob'] = finbert_probs[:, 2]
news_df['finbert_score'] = news_df['finbert_positive_prob'] - news_df['finbert_negative_prob']
news_df['finbert_sentiment_strength'] = np.maximum(news_df['finbert_positive_prob'], news_df['finbert_negative_prob'])

print("\nNews Data with Newly Generated FinBERT Probabilities & Scores Head:")
print(news_df.head())

# --- VERIFY SENTIMENT DISTRIBUTION ---
print("\n--- Sentiment Score Distribution ---")
plt.figure(figsize=(10, 6))
news_df['finbert_score'].hist(bins=50, edgecolor='black')
plt.title('Distribution of FinBERT Sentiment Scores (Re-analyzed)')
plt.xlabel('Sentiment Score (-1: Negative, 0: Neutral, 1: Positive)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
print(news_df['finbert_score'].describe())
print(f"Percentage of positive scores: {(news_df[news_df['finbert_score'] > 0.1].shape[0] / news_df.shape[0]) * 100:.2f}%")
print(f"Percentage of negative scores: {(news_df[news_df['finbert_score'] < -0.1].shape[0] / news_df.shape[0]) * 100:.2f}%")
print(f"Percentage of neutral scores: {(news_df[(news_df['finbert_score'] >= -0.1) & (news_df['finbert_score'] <= 0.1)].shape[0] / news_df.shape[0]) * 100:.2f}%")

# --- 3. Merging Stock and Shifted Sentiment Data ---
print("\n--- 3. Merging Stock and Shifted Sentiment Data ---")
daily_sentiment = news_df.groupby(['Date', 'Symbol']).agg(
    shifted_finbert_score=('finbert_score', 'mean'),
    shifted_finbert_neutral_prob=('finbert_neutral_prob', 'mean'),
    shifted_finbert_pos_prob=('finbert_positive_prob', 'mean'),
    shifted_finbert_neg_prob=('finbert_negative_prob', 'mean'),
    shifted_sentiment_strength=('finbert_sentiment_strength', 'mean')
).reset_index()

daily_sentiment['Date'] = daily_sentiment['Date'] + pd.Timedelta(days=1)

merged_df = pd.merge(stock_df, daily_sentiment, on=['Date', 'Symbol'], how='left')

sentiment_cols = [col for col in merged_df.columns if 'shifted_finbert' in col or 'shifted_sentiment_strength' in col]
merged_df[sentiment_cols] = merged_df[sentiment_cols].fillna(0)

print("\nMerged Data with Shifted Sentiment Head:")
print(merged_df.head())
print("\nMerged Data Info:")
merged_df.info()

merged_df.to_csv(MERGED_DATA_SAVE_FILE, index=False)
print(f"\nMerged data with enhanced shifted sentiment saved to '{MERGED_DATA_SAVE_FILE}'")

# --- 4. Enhanced Feature Engineering (including rolling sentiment) ---
print("\n--- 4. Enhanced Feature Engineering ---")
def create_engineered_features(df):
    df = df.copy()
    df['Prev_Close_1'] = df.groupby('Symbol')['Close'].shift(1)
    df['Prev_Close_2'] = df.groupby('Symbol')['Close'].shift(2)
    df['Prev_Close_3'] = df.groupby('Symbol')['Close'].shift(3)

    df['Daily_Return'] = (df['Close'] - df['Prev_Close_1']) / df['Prev_Close_1']
    previous_volume = df.groupby('Symbol')['Volume'].shift(1)
    df['Volume_Change'] = (df['Volume'] - previous_volume) / previous_volume
    df['Volume_Change'] = df['Volume_Change'].replace([np.inf, -np.inf], 0)

    df['High_Low_Range'] = (df['High'] - df['Low']) / df['Open']

    sentiment_window_sizes = [3, 5, 7]
    for window in sentiment_window_sizes:
        df[f'rolling_sentiment_avg_{window}d'] = df.groupby('Symbol')['shifted_finbert_score'].transform(lambda x: x.rolling(window=window, min_periods=1).mean())
        df[f'rolling_sentiment_std_{window}d'] = df.groupby('Symbol')['shifted_finbert_score'].transform(lambda x: x.rolling(window=window, min_periods=1).std())
        df[f'rolling_sentiment_std_{window}d'] = df[f'rolling_sentiment_std_{window}d'].fillna(0)

    df = df.fillna(0)
    return df

merged_df_features = create_engineered_features(merged_df.copy())
print("\nMerged Data with New Features (including rolling sentiment) Head:")
print(merged_df_features.head())
merged_df_features.info()

# --- 5. Loop through Stocks, Train, Evaluate, and Save Models ---
print("\n--- 5. Training Models Per Stock ---")
unique_symbols = merged_df_features['Symbol'].unique()
overall_performance = []

for symbol in unique_symbols:
    print(f"\n--- Processing Stock: {symbol} ---")
    stock_data = merged_df_features[merged_df_features['Symbol'] == symbol].copy()

    features = [
        'Open', 'High', 'Low', 'Close', 'Volume',
        'shifted_finbert_score', 'shifted_finbert_neutral_prob',
        'shifted_finbert_pos_prob', 'shifted_finbert_neg_prob',
        'shifted_sentiment_strength',
        'Prev_Close_1', 'Prev_Close_2', 'Prev_Close_3',
        'Daily_Return', 'Volume_Change', 'High_Low_Range'
    ]
    for window in [3, 5, 7]:
        features.append(f'rolling_sentiment_avg_{window}d')
        features.append(f'rolling_sentiment_std_{window}d')

    target = 'Close'

    stock_data = stock_data.dropna(subset=features)

    if len(stock_data) < SEQUENCE_LENGTH + 1:
        print(f"Skipping {symbol}: Not enough data ({len(stock_data)} rows) for sequence creation with SEQUENCE_LENGTH={SEQUENCE_LENGTH}.")
        continue

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(stock_data[features])

    target_scaler = MinMaxScaler(feature_range=(0, 1))
    target_scaler.fit(stock_data[[target]])

    joblib.dump(scaler, os.path.join(SCALER_SAVE_DIR, f'{symbol}_feature_scaler.pkl'))
    joblib.dump(target_scaler, os.path.join(SCALER_SAVE_DIR, f'{symbol}_target_scaler.pkl'))
    print(f"Scalers for {symbol} saved to {SCALER_SAVE_DIR}")

    def create_sequences(data, sequence_length, feature_count, target_idx):
        X, y = [], []
        for i in range(len(data) - sequence_length):
            X.append(data[i:(i + sequence_length)])
            y.append(data[i + sequence_length, target_idx])
        return np.array(X), np.array(y)

    target_index_in_features = features.index(target)
    X, y = create_sequences(scaled_data, SEQUENCE_LENGTH, len(features), target_index_in_features)

    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    if X_train.shape[0] == 0 or X_test.shape[0] == 0:
        print(f"Skipping {symbol}: Not enough data to create train/test sets after sequencing.")
        continue

    print(f"Shape of X_train ({symbol}): {X_train.shape}")
    print(f"Shape of y_train ({symbol}): {y_train.shape}")
    print(f"Shape of X_test ({symbol}): {X_test.shape}")
    print(f"Shape of y_test ({symbol}): {y_test.shape}")

    # Build the Bidirectional GRU model
    model = tf.keras.models.Sequential([
        tf.keras.layers.Bidirectional(tf.keras.layers.GRU(units=100, return_sequences=True),
                                      input_shape=(X_train.shape[1], X_train.shape[2])),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Bidirectional(tf.keras.layers.GRU(units=100, return_sequences=False)),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(units=50),
        tf.keras.layers.Dense(units=1)
    ])

    model.compile(optimizer='adam', loss='mean_squared_error')

    # Define Early Stopping callback
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=EARLY_STOPPING_PATIENCE,
        mode='min',
        restore_best_weights=True,
        verbose=1
    )

    history = model.fit(X_train, y_train,
                        epochs=MAX_EPOCHS,
                        batch_size=BATCH_SIZE,
                        validation_data=(X_test, y_test),
                        callbacks=[early_stopping],
                        verbose=0)

    print(f"Model Training Complete for {symbol}. Training stopped at epoch {len(history.history['loss'])}.")

    model_file_path = os.path.join(MODEL_SAVE_DIR, f'{symbol}_bidirectional_gru_enhanced_sentiment_model.keras')
    model.save(model_file_path)
    print(f"Model for {symbol} saved to {model_file_path}")

    # Plot training and validation loss
    plt.figure(figsize=(10, 6))
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{symbol} Bidirectional GRU Model Loss Over Epochs (Re-analyzed Sentiment & Early Stopped)')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

    # --- 6. Prediction and Evaluation for Current Stock ---
    print(f"\n--- 6. Prediction and Evaluation for {symbol} ---")
    predictions = model.predict(X_test, batch_size=BATCH_SIZE)

    predictions_full_scale = np.zeros((predictions.shape[0], len(features)))
    predictions_full_scale[:, features.index(target)] = predictions[:, 0]
    predictions_actual = target_scaler.inverse_transform(predictions_full_scale)[:, features.index(target)]

    y_test_full_scale = np.zeros((y_test.shape[0], len(features)))
    y_test_full_scale[:, features.index(target)] = y_test
    y_test_actual = target_scaler.inverse_transform(y_test_full_scale)[:, features.index(target)]

    r_squared = r2_score(y_test_actual, predictions_actual)
    print(f"R-squared for {symbol}: {r_squared:.4f}")

    previous_actual_close_start_idx = len(stock_data) - len(X_test) - 1
    previous_actual_close_end_idx = len(stock_data) - 1
    previous_actual_close = stock_data['Close'].iloc[previous_actual_close_start_idx : previous_actual_close_end_idx].values

    min_len_for_direction = min(len(y_test_actual), len(predictions_actual), len(previous_actual_close))
    y_test_actual_for_direction = y_test_actual[:min_len_for_direction]
    predictions_actual_for_direction = predictions_actual[:min_len_for_direction]
    previous_actual_close_for_direction = previous_actual_close[:min_len_for_direction]

    actual_move = np.sign(y_test_actual_for_direction - previous_actual_close_for_direction)
    predicted_move = np.sign(predictions_actual_for_direction - previous_actual_close_for_direction)

    correct_directions = np.sum(actual_move == predicted_move)
    total_predictions_for_direction = len(actual_move)
    directional_accuracy = correct_directions / total_predictions_for_direction

    print(f"Directional Accuracy for {symbol}: {directional_accuracy:.4f}")

    overall_performance.append({
        'Symbol': symbol,
        'R_squared': r_squared,
        'Directional_Accuracy': directional_accuracy,
        'Test_Samples': len(y_test_actual)
    })

print("\n--- Overall Model Performance Summary ---")
overall_performance_df = pd.DataFrame(overall_performance)
print(overall_performance_df)
```
